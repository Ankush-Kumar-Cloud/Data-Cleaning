# Ordinal Encoding

### by sklearn

```python
import pandas as pd
```

```python
df = pd.DataFrame({"size" : ["S","M","L","XL","S","M","L","S","S","L","XL","M"]})
```

```python
df
```

```python
ord_data = [["S","M","L","XL"]]
```

## import siketlearn

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)
oe.fit(df[["size"]])
```

```python
df["size_en"]=oe.transform(df[["size"]])
```

```python
df
```

## Ordinal Encoding

## map Function

```python
ord_data1 = {"S":0,"M":1,"XL":3,"L":2}
```

```python
df["size_en_map"]=df["size"].map(ord_data1)
```

```python
df
```
